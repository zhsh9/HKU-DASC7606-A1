```python
import matplotlib.pyplot as plt

def draw_loss_curve(losses, title='Loss Function Over Time', fs=(10, 5), filename='model', y_lim=None):
    # Set the size of the plot
    plt.figure(figsize=fs)
    
    # Plot the loss function as a line chart
    plt.plot(losses, marker='o', linestyle='-', color='red')
    
    # Add a title and labels
    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    
    # Set x-axis ticks to only show integers
    plt.xticks(range(len(losses)))

    # Optionally set the y-axis limits if provided
    if y_lim:
        plt.ylim(y_lim)
    
    # Optionally: add grid lines for better readability
    plt.grid(True)

    # Save the picture locally
    save_path = f'./losses/{filename}.png'
    plt.savefig(save_path)
    
    # Display the plot
    plt.show()
```

```python
import matplotlib.pyplot as plt

def draw_loss_curves(losses_list, labels=None, title='Loss Functions Over Time', fs=(10, 5), filename='model', y_lim=None):
    # Set the size of the plot
    plt.figure(figsize=fs)
    
    # Check if a single loss list is provided instead of a list of lists
    if not isinstance(losses_list[0], list):
        losses_list = [losses_list]  # Wrap it in a list if it's not a list of lists
        labels = [labels] if labels else ['Loss']  # Make sure labels is a list

    # Plot each set of loss values
    for losses, label in zip(losses_list, labels or [''] * len(losses_list)):
        plt.plot(losses, marker='o', linestyle='-', label=label)
    
    # Add a title and axis labels
    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    
    # Set x-axis ticks to show integer epochs
    plt.xticks(range(max(len(losses) for losses in losses_list)))

    # Optionally set the y-axis limits if provided
    if y_lim:
        plt.ylim(y_lim)
    
    # Add grid lines for better readability
    plt.grid(True)
    
    # Add a legend if any labels were provided
    if labels:
        plt.legend()
    
    # Save the plot as a PNG file
    plt.savefig(f'./losses/{filename}.png')
    
    # Display the plot
    plt.show()

# Example usage:
# draw_loss_curves([[0.8, 0.6, 0.4], [0.7, 0.5, 0.3]], labels=['Loss A', 'Loss B'])
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

def draw_loss_curve_with_break(loss, title, fs, filename):
    # Create a figure with two subplots sharing the x-axis
    fig = plt.figure(figsize=fs)
    gs = GridSpec(2, 1, height_ratios=[1, 3], hspace=0.05)  # Adjust the ratio based on your data
    
    # Upper subplot for the range 192000 to 193000
    ax1 = fig.add_subplot(gs[0])
    ax1.plot(loss, marker='o', linestyle='-')
    ax1.set_ylim(192000, 192600)  # Set the y-axis limits for the upper subplot
    
    # Lower subplot for the range 0 to 600
    ax2 = fig.add_subplot(gs[1], sharex=ax1)
    ax2.plot(loss, marker='o', linestyle='-')
    ax2.set_ylim(0, 600)  # Set the y-axis limits for the lower subplot
    
    # Hide the spines between ax1 and ax2
    ax1.spines['bottom'].set_visible(False)
    ax2.spines['top'].set_visible(False)
    ax1.xaxis.tick_top()
    ax1.tick_params(labeltop=False)  # Don't show the top tick labels
    ax2.xaxis.tick_bottom()

    # Add diagonal lines to indicate the break in axis
    d = .015  # Size of the diagonal lines
    kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
    ax1.plot((-d, +d), (-d, +d), **kwargs)        # Top-left diagonal
    ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # Top-right diagonal

    kwargs.update(transform=ax2.transAxes)  # Switch to the bottom axes
    ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # Bottom-left diagonal
    ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # Bottom-right diagonal

    # Set the title and labels
    ax1.set_title(title)
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')

    # Display the plot
    plt.show()

    # Save the plot as a PNG file
    fig.savefig(f'./losses/{filename}.png')

# Example usage:
# Assuming 'model7' is a list of loss values
draw_loss_curve_with_break(
    model7,
    'final model training: seed(3407), epochs(20), depth(18), lr(1e-3)',
    (20, 10),
    'model7'
)
```
